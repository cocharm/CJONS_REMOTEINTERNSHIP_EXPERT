## 1. Anscombe's quartet 데이터 셋 분석

```python
import pandas as pd
import seaborn as sns
```

```python
#(1) Anscombe's quartet 데이터 셋을 이용
aq = pd.read_excel('Desktop/AnscombesQuartet.xlsx', header=1)
aq
```

```python
# (2) 각 데이터 셋에 대한 요약통계 확인
aq.describe()
```

```python
# (3) 각 데이터 셋에 대한 상관계수 확인
corr = aq.corr(method='pearson')
corr
```

```python
sns.heatmap(corr)
```

## 2. physicalFitness 데이터 탐색적 분석 연습

```python
# (1)"physicalFitness.csv" 파일 로드
data = pd.read_csv('Desktop/physicalFitness.csv')
data.head()
```

```python
# (2)남녀 데이터를 분리
gr = data.groupby('sex')
m = gr.get_group(1)
f = gr.get_group(2)
```

```python
m
```

```python
f
```

```python
# (3)남녀 요약통계를 확인
c = data.columns
data.groupby('sex')[c].agg(['median', 'mean', 'std'])
```

```python
# (4)전체 데이터의 키와 몸무게의 상관관계 확인
p_corr = data[['height', 'weight']].corr(method='pearson')
p_corr
```

```python
sns.heatmap(p_corr)
```

```python
# (5)성별에 대한 연령별 평균 BMI 테이블 재구성하여 요약통계 확인
data.pivot_table(index='ageGroup', columns='sex', values='BMI').describe()
```
