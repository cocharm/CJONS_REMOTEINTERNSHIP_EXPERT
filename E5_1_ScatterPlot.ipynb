```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib 
```

```python
total = pd.read_csv('fa_weekend_total.csv',encoding='cp949')
total.head()
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
fig, (ax1, ax2,ax3,) = plt.subplots(3, 1, figsize = (10,20))

total = total[total['fa1_rct_n'] < 100]
total = total[total['fa2_rct_n'] < 125]

sns.scatterplot(data = total, x = 'fa1_rct_n', y = 'fa2_rct_n', s = 14, ax = ax1)

sns.scatterplot(data = total, x = 'fa1_rct_n', y = 'weekend_ratio', s = 14, ax = ax2)

sns.scatterplot(data = total, x = 'fa2_rct_n', y = 'weekend_ratio', s = 14, ax = ax3)

ax1.set_title('fa1_rct_n, fa2_rct_n 분포', fontsize = 16)
ax2.set_title('fa1_rct_n, weekend_ratio 분포', fontsize = 16)
ax3.set_title('fa2_rct_n, weekend_ratio 분포', fontsize = 16)

plt.show()
```
